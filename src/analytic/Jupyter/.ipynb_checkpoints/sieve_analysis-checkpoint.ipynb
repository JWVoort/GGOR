# Doorlatenheid van een korrelverdeling

Karman-Kozeny geeft de doorlatendheid van een poreus medium met uniforme korrel dat een bepaalde porositeit bezit. Deze doorlatendheid is gebaseerd op viskeuze stroming in poriÃ«n het het totale korreloppervlak binnen een eenheid van het poreuze medium. Een korrelverdeling heeft echter een bandbreedte aan korrels. Het totale oppervlak daarvan binnen een volume-eenheid kan daaruit worden afgeleid.

Het oppervlak van $n$ bolvormige korrels met diameter $d$ bedraagt

$$S=n\pi d^{2}$$

Om het totale oppervlak van een volume-eenheid van een monster te berekenen hoeven we alleen nog te wegen hoeveel van korrels van elke diameter in het monster aanwezig zijn.

De korrelverdeling die we door zeven bepalen is gebaseerd op het gewicht van de fractie die juist wel of niet door de gaten van een zeef passen. Omrekenen naar volume aan korrels geschiedt door deling door hun dichtheid, waarvoor gemakshalve een uniforme waarde wordt aangenomen, namelijk van zand, $\rho=2.64g/cm^{3}$.

Het korreloppervlak van een monster met gewicht $\delta p$ waarin $n$ korrels aanwezig zijn met korreldiameter $d$ is

$$\delta s=n\pi d^{2}$$

de massa van deze $n$ korrels bedraagt

$$\delta p=\rho n\frac{1}{6}\pi d^{3}$$

waaruit het aantal korrels $n$ in deze massa $\delta p$ volgt

$$n=6\frac{\delta p}{\rho\pi d^{3}}$$

zodat, via eleminatie van $n$ de verhouding tussen massa en oppervlak wordt verkregen

$$\delta s=6\frac{\pi d^{2}}{\rho\pi d^{3}}\delta p$$

Dus

$$\delta s=\frac{6}{d}\frac {\delta p }{\rho}$$

Voor de gehele korrelverdelingscurve geldt dan

$$ S = \frac 6 \rho \sum_i\frac {\delta \rho_i} {d_i} $$


en voor uniforme korrels geldt natuurlijk ook

$$ S = \frac 6 d_u \frac {\sum_i \delta p_i} \rho$$


Omgekeerd volgt hieruit de effective korreldiameter

$$ d_u = \frac 6 S \frac {\sum \delta p_i} \rho $$

Waarbij $d_u$ zodanig is dat het bij gelijk gewicht als materiaal uit de korreldiameter het oppervlak dezelfde is. Dit resultaat heeft dus niets met porositeit te maken. Het vertelt alleen welke korreldiameter zorgt bij een gelijk gewicht als het uitgangsmateriaal voor een gelijk gewicht.

$\sum_i \delta \rho_i = p = \rho (1 - \epsilon) V$ met $V$ het volume waarin het gewicht $p$ aan korrels zit. Het aanwezige korreloppervlak is dus ook gelijk aan

$$S= \frac 6 d_u (1 - \epsilon) V$$

Omgekeerd kan uit een bekend oppervlak in een gegeven vollume de effectieve korreldiameter worden berekend
$$ d_u = 6 (1 - \epsilon) \frac V S$$

```python
import os
import numpy as np
from scipy.special import expi
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Sieve curve
pth = '/Users/Theo/Instituten-Groepen-Overleggen/TU/Studenten/'+\
        'Promovendi_eigen/Marloes van Ginkel/Promotie/sieve_analysis/'
sc = pd.read_excel(os.path.join(pth,'Sievetests_theo.xlsx'), sheet_name='Sieve tests', header=3,
                  index_col=0)
sc['test'] = sc['Mean'] * 0
sc['test'].loc[sc['MVII1'] > 4] = 100.

sc.index = np.array(sc.index) / 1000
for col in sc.columns:
    sc[col] /= 100.
```

```python
fig, ax = plt.subplots()
ax.set_title('sieve curves')
ax.set_xlabel('d [mm]')
ax.set_ylabel('p')
ax.set_xscale('log')
ax.grid()

for col in sc.columns:
    ax.plot(sc.index, sc[col], label=col)

    ax.legend()
ax.plot(sc.index, sc['Mean'], 'k--', lw=2, label=col)

ax.plot()
```

```python
def area(sc, rho=2640):
    '''Return the surface area of the grains under each sieve curve [m2/kg]
    parameters
    ----------
        sc: pd.DataFrame
            index are the sieve sizes in mm
            the valuea are the p values on the sieves, total is assumed 100% = 1 kg
        rho: float
            grain density in kg/m3
    '''
    sv = np.array(sc.index) # sieve size
    d = np.logspace(np.log10(sv[0]), np.log10(sv[-1]), 1001) # interpolation points between the sieves
    plt.title('sieve curves')
    plt.xscale('log')
    plt.xlabel('d [m]')
    plt.ylabel('p')
    plt.grid()
    S=dict() # dict to store the results for each sieve curve
    deff = dict()
    for col in sc.columns:
        p = np.interp(d, sv, sc[col].values)
        
        plt.plot(d, p, label=col)

        dm = 0.5 * (d[:-1] + d[1:]) # grain size between successive d values
        dp = np.diff(p) # weight between successive diameters
        ds = 6 * dp / rho / dm # surface area of this dp
        S[col] = np.sum(ds) # total grain surface area under the sieve curve 
        
        # add the grains below the first sieve
        S[col] += 6 * p[0] / rho / d[0]
        deff[col] = 6 / S[col] * sum(dp) /rho
        
        plt.plot(deff[col], np.interp(deff[col], d, p), '.')
    plt.legend()
    plt.show()
    return S, deff
```

```python
S, deff = area(sc)
for k in S:
    print('{:6} : {:8.1f} m2/kg, d_eff = {:.3g} mm'.format(k, S[k], 1000 * deff[k]))
```
